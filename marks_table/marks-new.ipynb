### BIRCH

threshold - Радиус подкластера меньше данного значения
branching_factor - Максимальное число CF в каждой вершине
n_clusters - Количество кластеров после последнего шага кластеризации, на котором подкластеры из листьев рассматриваются как новые выборки. Если установлено значение None, последний этап кластеризации не выполняется, и подкластеры возвращаются как есть

```python
brc = Birch(branching_factor=50, n_clusters=None, threshold=1.5)
brc.fit(table.values)
```

Predict -  чтобы получить список точек и их кластер

```python
labels = brc.predict(table.values)
len(labels)
```

Строим график, для разных кластеров - разные цвета

```python
plt.scatter(table.values[:,0], table.values[:,1], c=labels, cmap='rainbow', alpha=0.7, edgecolors='b')
```
